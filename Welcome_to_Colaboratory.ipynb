```
import nltk
from nltk.corpus import brown
from nltk.tag import hmm
import matplotlib.pyplot as plt

import numpy as np
import seaborn as sns
from collections import defaultdict
from sklearn.model_selection import KFold
from sklearn.metrics import confusion_matrix

class HMMTrainer:
    def __init__(self):
        self.states = set()  # Set of POS tags
        self.vocab = set()   # Set of words in the vocabulary
        self.transition_probs = defaultdict(lambda: defaultdict(float))
        self.emission_probs = defaultdict(lambda: defaultdict(float))
        self.initial_state_probs = defaultdict(float)

    def train(self, train_data):
        # Data preprocessing and initialization of parameters
        self.initialize_parameters(train_data)

        # Training: Estimate transition and emission probabilities
        self.estimate_transition_probabilities(train_data)
        self.estimate_emission_probabilities(train_data)

    def initialize_parameters(self, train_data):
        # Initialize self.states, self.vocab, and other parameters
        for sentence in train_data:
            for word, tag in sentence:
                self.states.add(tag)
                self.vocab.add(word)

        # Initialize transition_probs, emission_probs, and initial_state_probs
        num_states = len(self.states)
        for state1 in self.states:
            for state2 in self.states:
                self.transition_probs[state1][state2] = 1.0 / num_states
            self.initial_state_probs[state1] = 1.0 / num_states

    def estimate_transition_probabilities(self, train_data):
        # Estimate transition probabilities from train_data
        transition_counts = defaultdict(lambda: defaultdict(int))

        for sentence in train_data:
            prev_state = None
            for _, tag in sentence:
                if prev_state is not None:
                    transition_counts[prev_state][tag] += 1
                prev_state = tag

        for state1 in self.states:
            total_transitions = sum(transition_counts[state1].values())
            for state2 in self.states:
                self.transition_probs[state1][state2] = transition_counts[state1][state2] / total_transitions

    def estimate_emission_probabilities(self, train_data):
        # Estimate emission probabilities from train_data
        emission_counts = defaultdict(lambda: defaultdict(int))

        for sentence in train_data:
            for word, tag in sentence:
                emission_counts[tag][word] += 1

        for state in self.states:
            total_emissions = sum(emission_counts[state].values())
            for word in self.vocab:
                self.emission_probs[state][word] = emission_counts[state][word] / total_emissions

    def viterbi(self, sentence):
      n = len(sentence)
      dp = [defaultdict(float) for _ in range(n)]
      backpointer = [defaultdict(str) for _ in range(n)]  # Initialize as defaultdict(str)

      # Initialization step
      for state in self.states:
          dp[0][state] = self.initial_state_probs[state] * self.emission_probs[state][sentence[0]]

      # Recursion step
      for t in range(1, n):
          for state1 in self.states:
              for state2 in self.states:
                  prob = dp[t - 1][state2] * self.transition_probs[state2][state1] * self.emission_probs[state1][sentence[t]]
                  if prob > dp[t][state1]:
                      dp[t][state1] = prob
                      backpointer[t][state1] = state2

      # Termination step
      best_last_state = max(self.states, key=lambda state: dp[n - 1][state])
      best_tag_sequence = [best_last_state]

      # Backtrack to find the best tag sequence
      for t in range(n - 1, 0, -1):
          best_last_state = backpointer[t][best_last_state]
          best_tag_sequence.insert(0, best_last_state)

      return best_tag_sequence

    def evaluate_per_pos_accuracy(self, true_tags, predicted_tags):
        # Calculate per POS accuracy
        per_pos_accuracy = defaultdict(lambda: {"correct": 0, "total": 0})

        for true_tag, predicted_tag in zip(true_tags, predicted_tags):
            per_pos_accuracy[true_tag]["total"] += 1
            if true_tag == predicted_tag:
                per_pos_accuracy[true_tag]["correct"] += 1

        for pos, accuracy_data in per_pos_accuracy.items():
            accuracy = accuracy_data["correct"] / accuracy_data["total"] if accuracy_data["total"] > 0 else 0
            print(f"Accuracy for POS '{pos}': {accuracy:.2%}")

    def evaluate(self, test_data):
        # Evaluate the trained HMM on test_data
        correct = 0
        total = 0

        for sentence in test_data:
            words, true_tags = zip(*sentence)
            predicted_tags = self.viterbi(words)

            correct += sum(1 for true, predicted in zip(true_tags, predicted_tags) if true == predicted)
            total += len(true_tags)

        accuracy = correct / total
        return accuracy

    def evaluate_tags(self, test_data):
      # Evaluate the trained HMM on test_data
      true_tags = []
      predicted_tags = []
      for sentence in test_data:
          words, true_tags_sentence = zip(*sentence)
          predicted_tags_sentence = self.viterbi(words)
          true_tags.extend(true_tags_sentence)
          predicted_tags.extend(predicted_tags_sentence)
      return true_tags, predicted_tags

nltk.download('brown')
nltk.download('universal_tagset')

# Split the corpus into sentences
sentences = brown.sents()
tagged_sentences = brown.tagged_sents(tagset='universal')

kf = KFold(n_splits=5, shuffle=True, random_state=42)

accuracies = []

# Perform 5-fold cross-validation
for train_index, test_index in kf.split(tagged_sentences):
    train_set = [tagged_sentences[i] for i in train_index]
    test_set = [tagged_sentences[i] for i in test_index]

    # Train the Hidden Markov Model (HMM) POS tagger
    trainer = HMMTrainer()
    pos_tagger = trainer.train(train_set)

    # Evaluate the tagger on the test set
    accuracy = trainer.evaluate(test_set)
    accuracies.append(accuracy)

# Calculate and print the average accuracy over 5 folds
average_accuracy = np.mean(accuracies)
print("Average Accuracy:", average_accuracy)

# Split data into training and testing sets
train_size = int(0.8 * len(tagged_sentences))
train_set = tagged_sentences[:train_size]
test_set = tagged_sentences[train_size:]

# Train the Hidden Markov Model (HMM) POS tagger
trainer = HMMTrainer()
pos_tagger = trainer.train(train_set)

# Evaluate the tagger on the test set
accuracy = trainer.evaluate(test_set)
print("Accuracy:", accuracy)

# Example sentences to tag
true_tags, predicted_tags = trainer.evaluate_tags(test_set)
trainer.evaluate_per_pos_accuracy(true_tags, predicted_tags)
```

```
all_tags = sorted(set(true_tags+predicted_tags))
confusion = confusion_matrix(true_tags,predicted_tags, labels= all_tags, normalize='true')

plt.figure(figsize=(12, 8))
sns.set(font_scale=1.2)  # Adjust font size if needed
sns.heatmap(confusion, annot=True, fmt=".2f", cmap="Blues", xticklabels=all_tags, yticklabels=all_tags)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.show()
```
